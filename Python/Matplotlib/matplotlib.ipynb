```python
import matplotlib.pyplot as plt


# a simple plot
X = [1, 2, 3]
Y = [4, 5, 6]
plt.plot(X, Y)
plt.show()
```

```python
# Movies Data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]

plt.plot(oscar_years, oscar_revenue)
plt.xlabel("Years")
plt.ylabel("Revenue (millions)")
plt.title("Revenue of Oscar Winner Movies")
```

```python
# Plot multiple
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275]

plt.plot(oscar_years, oscar_revenue)
plt.plot(non_oscar_years, non_oscar_revenue)
plt.xlabel("Years")
plt.ylabel("Revenue (millions)")
plt.title("Revenue of Oscar Winner Movies")
plt.legend(["Oscar Winner Movies", "Non-Oscar Winner Movies"])
plt.grid(True)
plt.tight_layout()
```

```python
plt.plot(oscar_years, oscar_revenue, "o-b", label="Oscar Winners")
plt.plot(non_oscar_years, non_oscar_revenue, ">--g", label="Non-Oscar Winners")

plt.plot(oscar_years, oscar_revenue, color="black", linestyle="-", marker="o",label="Oscar Winners")
plt.plot(non_oscar_years, non_oscar_revenue, color="red", linestyle="--", marker=">", linewidth=5, label="Non-Oscar Winners")
```

```python
# Bar charts
movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
revenue = [1005, 170, 427, 133, 232]

plt.bar(years, revenue)
plt.xlabel("Years")
plt.ylabel("Revenue (millions)")
plt.title("Revenue of Oscar Winner Movies")

# Add labels on top of bars

for i in range(len(revenue)):
  plt.text(years[i], revenue[i]+20, str(revenue[i]), ha='center')

plt.ylim(0, max(revenue)+200)



```

```python
# comparing data with bar charts
oscar_movies = ["The Dark Knight","The Hurt Locker","The King's Speech","The Artist","Argo"]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

years = [2008, 2009, 2010, 2011, 2012]

# Plotting Bar Chart
import numpy as np
x= np.arange(len(years))
width = 0.4

plt.bar(x-width/2, oscar_revenue, width, label="Oscar Winners")
plt.bar(x+width/2, non_oscar_revenue, width, label="Non-Oscar Winners")

plt.xlabel("Years")
plt.ylabel("Revenue (millions)")
plt.title("Revenue of Oscar Winner Movies")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(x, years)
```

```python
# Scatter Plots

people = ["Person A", "Person B", "Person C", "Person D", "Person E",
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
blood_pressure = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

plt.title("Age vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
sizes = blood_pressure
colors = ["green" if bp<135 else "red" for bp in blood_pressure]
plt.scatter(age, blood_pressure, c=colors, alpha =0.5, s=sizes)

plt.scatter(age, blood_pressure, cmap = "orRd")
plt.colorbar(label="Blood Pressure")

for i in range(len(age)):
    plt.annotate(people[i], xy=(age[i], blood_pressure[i]))
plt.xlim(min(age), max(age)+10)
```

```python
# Multiple scatter plots
cities = ["City A", "City B", "City C", "City D", "City E"]

# Winter: Temperature (°C) vs Humidity (%)
winter_temp = [5, 2, 10, 0, 7]
winter_humidity = [80, 75, 65, 85, 70]

# Summer: Temperature (°C) vs Humidity (%)
summer_temp = [25, 30, 28, 35, 27]
summer_humidity = [60, 50, 55, 45, 65]

# Plots
plt.scatter(winter_temp, winter_humidity, color="blue", label="Winter")
plt.scatter(summer_temp, summer_humidity, color="orange", label="Summer")

plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")
plt.legend()
```

```python
# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

plt.pie(amounts, labels=expenses)

# set style
plt.style.use("ggplot")

# add %-age of each slice
plt.pie(amounts, labels=expenses,  autopct='%1.1f%%')

#set colors
colors = ["purple", "green", "blue", "pink", "yellow"]
plt.pie(amounts, labels=expenses,  autopct='%1.1f%%', colors=colors)

# wedgeprops
plt.pie(amounts, labels=expenses,  autopct='%1.1f%%',
        wedgeprops={"edgecolor":"black", "linewidth": 1, "linestyle":"-.", "fill":False}
       )
```
